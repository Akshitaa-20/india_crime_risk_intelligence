```python
import pandas as pd
df = pd.read_csv("/content/Crime against Women (2001-2022).csv")
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.head(3)
```

```python
df.rename(columns={'CRIME HEAD': 'Crime_Type'}, inplace=True)
```

```python
df_long = df.melt(
    id_vars='Crime_Type',
    var_name='Year',
    value_name='Crime_Count'
)

df_long.head()
```

```python
df_long['Year'] = df_long['Year'].astype(int)
df_long['Crime_Count'] = pd.to_numeric(df_long['Crime_Count'], errors='coerce')

df_long.info()
```

```python
df_long.isnull().sum()
```

```python
df_long = df_long.dropna()
```

### Data Reshaping

- Converted dataset from wide to long format
- Each row now represents a crime type in a specific year
- Data spans from 2001 to 2022
- Dataset is now suitable for analysis and ML

```python
df_long.to_csv("/content/cleaned_crime_data.csv", index=False)
```
