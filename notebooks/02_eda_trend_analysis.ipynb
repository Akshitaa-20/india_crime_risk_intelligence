```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Load cleaned dataset
df = pd.read_csv("/content/cleaned_crime_data.csv")
df.head()
```

# Overall Crime Trend

```python
yearly_trend = df.groupby('Year')['Crime_Count'].sum().reset_index()
plt.figure(figsize=(10,5))
plt.plot(yearly_trend['Year'], yearly_trend['Crime_Count'])
plt.title("Total Crimes Against Women in India (2001â€“2022)")
plt.xlabel("Year")
plt.ylabel("Number of Crimes")
plt.grid(True)
plt.show()
```

# Top Crime Types

```python
top_crimes = (
    df.groupby('Crime_Type')['Crime_Count']
      .sum()
      .sort_values(ascending=False)
      .head(5)
)
top_crimes.plot(kind='bar', figsize=(8,5))
plt.title("Top 5 Crime Types Against Women (2001â€“2022)")
plt.ylabel("Total Cases")
plt.show()
```

# Crime Type Trends Over Time

```python
major_crimes = df['Crime_Type'].value_counts().head(5).index
filtered_df = df[df['Crime_Type'].isin(major_crimes)]
plt.figure(figsize=(12,6))
sns.lineplot(data=filtered_df, x='Year', y='Crime_Count', hue='Crime_Type')
plt.title("Trends of Major Crime Types Over Time")
plt.show()
```

# Growth Rate Analysis

```python
yearly_trend['Growth_Rate'] = yearly_trend['Crime_Count'].pct_change() * 100
plt.figure(figsize=(10,5))
plt.plot(yearly_trend['Year'], yearly_trend['Growth_Rate'])
plt.title("Year-on-Year Growth Rate (%)")
plt.ylabel("Growth %")
plt.axhline(0)
plt.show()
```

### Key EDA Insights

- The overall number of reported crimes against women shows a long-term increasing trend from 2001 to 2022.
- A few crime categories consistently account for a majority of total reported crimes across the years.
- Certain years exhibit sharp deviations in crime counts, indicating the influence of external socio-political or legal factors.
- The COVID-19 period (2020â€“2021) shows a noticeable disruption in reporting trends, likely due to lockdowns and reduced access to reporting mechanisms.
- Recent years indicate a renewed upward trend in crime counts, highlighting the need for continued monitoring and intervention.
